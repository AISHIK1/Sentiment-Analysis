```python
def read_file(filepath):
    with open(filepath) as f:
        str_text=f.read()
    return str_text
```

```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm',disable=['parser','tagger','ner'])
```

```python
nlp.max_length=119863
```

```python
def seperate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
d=read_file('moby_dick_four_chapters.txt')
```

```python
token=seperate_punc(d)
```

```python
len(token)
```

```python
train_len=25+1
text_sequences=[]
for i in range(train_len,len(token)):
    seq=token[i-train_len:i]
    text_sequences.append(seq)
```

```python
print(text_sequences[0])
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences=tokenizer.texts_to_sequences(text_sequences)
```

```python
y=to_categorical(y,num_classes=vocabulary_size+1)
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
def create_model(vocabulary_size,seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, 25, input_length=seq_len))
    model.add(LSTM(150, return_sequences=True))
    model.add(LSTM(150))
    model.add(Dense(150, activation='relu'))

    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model
```

```python
model=create_model(vocabulary_size+1,seq_len)
```

```python
from pickle import dump,load
```

```python
model.fit(x,y,batch_size=128,epochs=300,verbose=1)
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):  
  
    output_text = []
    input_text = seed_text

    for i in range(num_gen_words):
        
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]
        pred_word = tokenizer.index_word[pred_word_ind] 
        
        output_text.append(pred_word)
    return ' '.join(output_text)
```

```python
import random
random.seed(101)
random_pick=random.randint(1,len(text_sequences))
random_seed_text=text_sequences[random_pick]
seed_text=' '.join(random_seed_text)
```

```python
generate_text(model,tokenizer,seq_len,seed_text,num_gen_words=25)
```
